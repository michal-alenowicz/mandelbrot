```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#nowe funkcje do "zoomowania" macierzy

#funkcja przyjmuje macierz o rozmiarze 1200x1200 z LICZBAMI RZECZYWISTYMI, zwraca macierz o tym samym rozmiarze
#złożoną z 600 środkowych wierszy i 600 środkowych kolumn oryginału poprzetykanych wierszami i kolmunami zer
#UWAGA 1200-ta kolumna to zera. 1200-ty wiersz to zera

def insert_zeros(M):  #ta funkcja zostawia dodane zera w spokoju. pierwsza kolejna iteracja k powinna je wytępić
    M_cut = M[300:900,300:900]
#    print (M_cut.shape)
#    print(M_cut)
    A = np.zeros((1200,1200))
    A[::2,::2] = M_cut
    return (A)


def insert_zeros_real_mean(M):    #ta funkcja wypełnia dodane zera średnimi
    M_cut = M[300:901,300:901]
#    print (M_cut.shape)
#    print(M_cut)
    A = np.zeros((1202,1202))
    A[::2,::2] = M_cut
    print(A.shape)
    for i in range(1201):
        for j in range(1201):
            if (i%2 == 0) and (j%2 != 0):
                A[i][j] = (A[i][j-1]+A[i][j+1])/2
            elif (i%2 != 0) and (j%2 == 0):
                A[i][j] = (A[i-1][j]+A[i+1][j])/2
            elif (i%2 != 0) and (j%2 != 0):
                A[i][j] = (A[i-1][j-1]+A[i-1][j+1]+A[i+1][j-1]+A[i+1][j+1])/4
            else:
                pass
    A = A[:1200,:1200]
    print(f'kształt A: {A.shape}')
    return (A)


#UWAGA 1200-ta kolumna to zera. 1200-ty wiersz to zera - dlatego do wyliczania średnich będzie brana macierz 1201x1201,
#a potem przycięta do 1200x1200
def insert_zeros_complex_mean(M):
    M_cut = M[300:901,300:901]
    print (M_cut.shape)
    A = np.zeros((1202,1202),dtype=np.complex128)
    A[::2,::2] = M_cut
    print (A)
    print(A.shape)
    for i in range(1201):
        for j in range(1201):
            if (i%2 == 0) and (j%2 != 0):
                A[i][j] = (A[i][j-1]+A[i][j+1])/2
            elif (i%2 != 0) and (j%2 == 0):
                A[i][j] = (A[i-1][j]+A[i+1][j])/2
            elif (i%2 != 0) and (j%2 != 0):
                A[i][j] = (A[i-1][j-1]+A[i-1][j+1]+A[i+1][j-1]+A[i+1][j+1])/4
            else:
                pass
    A = A[:1200,:1200]
    print(f'kształt A: {A.shape}')
    return (A)
    

def NaN_exterminator(M):
    M = np.nan_to_num(M)  #powinno zastąpić każde nan zerem, każde inf wysokim floatem?
    print(f'NaN exterminator pozostawił {np.count_nonzero(np.isnan(M))} nanow')
    return (M)



#POPRAWIONY NIEOBRÓCONY - startowy wycinek dziedziny - potem będzie zoomowane w jego środek
x_range = np.linspace(-1,-0.3,1200)
y_range = np.linspace(0,-0.7,1200)

C = np.zeros((1200,1200),dtype=np.complex128)

for i in range(1200):
    for j in range(1200):
        C[i][j]=complex(x_range[j],y_range[i])
        
Z = np.zeros((1200,1200),dtype=np.complex128)
K = np.zeros((1200,1200))

plt.figure(figsize=(11,10),dpi=100)
plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)

for zoom in range (1,6): #będzie 5 poziomów "zooma" macirzy, na kazdym pozostaje 25% punktów z poprzedniego
    for k in range(1,21):  #na każdym poziomie "zooma" macierzy będzie k od 1 do 10 włącznie
        print(f'zoom={zoom}, k={k}, C_NaN = {np.count_nonzero(np.isnan(C))}, Z_NaN = {np.count_nonzero(np.isnan(Z))}')
        Z = Z**2 + C
        for i in range(1200):
            for j in range(1200):
                if (abs(Z[i][j])>2) and (K[i][j]==0):
                    K[i][j] = 100-(k+20*(zoom-1))
        plt.title(f'zoom level = {zoom}  k = {k}')
        sns.heatmap(K[15*k:1200-15*k, 15*k:1200-15*k],cmap='viridis',cbar=False)
        plt.savefig('mandel_frame'+str(k+20*(zoom-1))+'.png')
        Z = NaN_exterminator(Z)  #A NIECH USUWA nan PO KAŻDYM k!
    
    C = insert_zeros_complex_mean(C)
    Z = insert_zeros_complex_mean(Z)
    Z = NaN_exterminator(Z)  #i po rozcieńczaniu macierzy niech też eksterminuje nany
    K = insert_zeros_real_mean(K)  #UWAGA - tym razem niech uśrednia
    
#powinno to dać 20 obrazków na każdym z 5 poziomów zooma macierzy
```
