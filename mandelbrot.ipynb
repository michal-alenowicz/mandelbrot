```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#zwykła wersja - najpierw obliczenia, potem zapis jednego png
x_range = np.linspace(-2,1,1000)
y_range = np.linspace(1.5,-1.5,1000)

C = np.zeros((1000,1000),dtype=complex)

for i in range(1000):
    for j in range(1000):
        C[i][j]=complex(x_range[i],y_range[j])
        
Z = np.zeros((1000,1000),dtype=complex)
K = np.zeros((1000,1000))

for k in range(1,101):
    print(k)
    for i in range(1000):
        for j in range(1000):
            Z[i][j]=(Z[i][j])**2 + C[i][j]
            if (abs(Z[i][j])>2) and (K[i][j]==0):
                K[i][j] = 100-k
    
plt.figure(figsize=(11,10),dpi=100)
plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
sns.color_palette("viridis", as_cmap=True)
sns.heatmap(K)
plt.savefig('mandelbrot.png')
```

```python
plt.figure(figsize=(11,10),dpi=100)
plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
sns.heatmap(K,cmap='viridis',center=60)
plt.savefig('mandelbrot.png')
```
